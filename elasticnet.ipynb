```r
# testing to find best k value . Condsidering to use knn for now for imputation 

library(VIM)
library(dplyr)

# Picked a variable STRATEGY_MATH.5 with no NAs
var_to_test <- "STRATEGY_MATH.5"
true_values <- data[[var_to_test]]

# Randomly introduce NA in known places
set.seed(123)
test_indices <- sample(which(!is.na(true_values)), 50)
data_test <- data
data_test[[var_to_test]][test_indices] <- NA

#Try different k values and record RMSE
results <- data.frame(k = integer(), RMSE = numeric()) 

for (k in c(3, 5, 7, 9)) {
  imputed <- kNN(data_test, k = k, imp_var = FALSE)
  imputed_values <- imputed[[var_to_test]][test_indices]
  rmse <- sqrt(mean((imputed_values - true_values[test_indices])^2, na.rm = TRUE))
  results <- rbind(results, data.frame(k = k, RMSE = rmse))
}

print(results)

# k=9 Smallest RMSE. lets do imputation with this.

```

```r
# Load necessary packages
library(caret)
library(glmnet)
library(tidyverse)

# Function to do Elastic Net regression with CV
run_elasticnet_cv <- function(data, outcome_var, predictor_vars) {
    
  data <- data[, c(outcome_var, predictor_vars)]
  data <- na.omit(data)

  # Prepare predictors (x) and response (y)
  x <- model.matrix(as.formula(paste(outcome_var, "~ .")), data)[, -1]  # remove intercept
  y <- data[[outcome_var]]

  # Define CV control
  ctrl <- trainControl(method = "repeatedcv", number = 5, repeats = 3)
    
  tune_grid <- expand.grid(
  alpha = seq(0.1, 0.9, by = 0.2),                     # avoid alpha=0 or 1 extremes
  lambda = 10^seq(-3, 0.5, length.out = 30)            # avoid overly large lambda
)
  # Train Elastic Net model using caret
  set.seed(123)
    enet_model <- train(
  x = x,
  y = y,
  method = "glmnet",
  trControl = ctrl,
  tuneGrid = tune_grid
)

  # Print results
  cat("\n=====================================\n")
  cat("Elastic Net Regression with CV for:", outcome_var, "\n")
  cat("=====================================\n")
  print(enet_model)

  # Best tuning parameters
  best_alpha <- enet_model$bestTune$alpha
  best_lambda <- enet_model$bestTune$lambda
  final_model <- enet_model$finalModel
  cat("\nModel Evaluation:\n")
  cat("Best Alpha:", round(best_alpha, 3), "\n")
  cat("Best Lambda:", round(best_lambda, 5), "\n")
  cat("Best RMSE:", round(min(enet_model$results$RMSE), 3), "\n")
  cat("Best R-squared:", round(max(enet_model$results$Rsquared), 3), "\n")

  # Extract selected predictors
  coef_enet <- coef(final_model, s = best_lambda)
  selected_vars <- rownames(coef_enet)[which(coef_enet != 0)][-1]  # drop intercept
  cat("\nSelected Predictors:\n")
  print(selected_vars)

  # Fit final lm model using selected predictors
  if (length(selected_vars) > 0) {
    selected_clean_vars <- unique(gsub("`", "", gsub(":[^:]*$", "", selected_vars)))
    selected_clean_vars <- selected_clean_vars[selected_clean_vars %in% colnames(data)]
    final_formula <- as.formula(paste(outcome_var, "~", paste(selected_clean_vars, collapse = "+")))
    final_lm <- lm(final_formula, data = data)
    cat("\nSummary of Final Linear Model:\n")
    print(summary(final_lm))
  } else {
    final_formula <- as.formula(paste(outcome_var, "~ 1"))
    final_lm <- lm(final_formula, data = data)
    cat("\nNo predictors selected. Intercept-only model:\n")
    print(summary(final_lm))
  }

  # Return all relevant output
  return(list(
    model = enet_model,
    final_model = final_model,
    best_alpha = best_alpha,
    best_lambda = best_lambda,
    selected_vars = selected_vars,
    final_lm_model = final_lm
  ))
}

# Define target variables and predictor set
target_vars <- c("MATH5", "ENG5", "MATH9", "ENG9")
predictor_vars <- setdiff(names(data_imputed2), target_vars)

# Run for each target
enet_math5 <- run_elasticnet_cv(data_imputed2, "MATH5", predictor_vars)
enet_eng5  <- run_elasticnet_cv(data_imputed2, "ENG5", predictor_vars)
enet_math9 <- run_elasticnet_cv(data_imputed2, "MATH9", predictor_vars)
enet_eng9  <- run_elasticnet_cv(data_imputed2, "ENG9", predictor_vars)
```
